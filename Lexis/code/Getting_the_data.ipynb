```python
# This code creates a dictionary called documents, where the files are stored. Each document has a number as its index. 

# Initializing the dictionary
documents = {}
# Setting inicial value for the first document
i = 0

# Initializing for loop over all files in folder
for filename in os.listdir():
    # Initialize document
    doc = ''
    #Create a temporary file. Within this file we will get every paragraph 
    temp = Document('{}'.format(filename))

    for p in temp.paragraphs:
        print(p.text)
        print('###########')
        doc = doc + ' ' + p.text

    documents[i] = doc
    
    i+=1
    
    

    
    
```

```python
#! pip install nltk

# We are going to use the NLTK package for Natural language Processing. 

# For the first model we will use the word tokenizer. This means that we are going to treat words separately. The other choice
# would be sentence tokenizer. 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

# This is our own tokenizer
from nltk.tokenize import RegexpTokenizer
```

```python
#Variables

# This tokenizer eliminates 
tokenizer = RegexpTokenizer(r'\w+')



# https://regex101.com/r/cO8lqs/4
```

```python
def clean_text(file):
    clean_file = tokenizer.tokenize(file)
    clean_file = [x for x in clean_file if x not in stopwords.words()]
    
    final_file = []

    for item in clean_file:
        # This checks if the word has more than one letter
        if len(item) != 1:
            # This code eliminates numbers that are in string format
            try:
                int(item)

            except:
                final_file.append(item)

    return final_file
```

```python
stopwords.add('')
```

```python
# Now let's tokenize each document. this takes a while....
tokenized_documents = {}
i = 0
for item in documents.keys():
    print(i)
    
    i += 1
    
    tokenized_documents[item] = clean_text(documents[item])

```

```python
# Now let's construct the vocabulary used in all documents. when we do set(), we grab unique words in list. We loop through
# all tokenized documents and add each words. Then we eliminate the duplicates, constructing the vocab. This will be the
# columns of our TDI-DF matrix (That matrix that I've shown you with each document as each row).
vocab = []
for item in tokenized_documents.keys():
    vocab = vocab + tokenized_documents[item]

vocab = set(vocab)
print(len(vocab))
```
